**Q10**: Produce a list of facilities with a total revenue less than 1000.
The output of facility name and total revenue, sorted by revenue. Remember
that there's a different cost for guests and members!

```python
sql_stmt="""
SELECT DISTINCT F.name As fac_name, 
              (M.firstname||M.surname) As mem_name,
               F.membercost,
               F.guestcost,
               M.memid
                FROM Facilities As F
                Join Bookings As B
                On F.facid=B.facid
                Join Members As M
                On B.memid=M.memid 
                WHERE M.memid!='0'
"""
pd.read_sql(sql_stmt, conn).head(50)
```

```python
sql_stmt="""
With join_guest As (
    SELECT F.name As fac_name,
            F.membercost,
            F.guestcost, 
            (M.firstname||M.surname) As mem_name,
            F.facid
        FROM Facilities As F
        Join Bookings As B
        On F.facid=B.facid
        Join Members As M
        On B.memid=M.memid
        Where M.memid='0'
        ),
join_mem As (
    SELECT DISTINCT F.name As fac_name,
                    F.membercost,
                    F.guestcost, 
                    (M.firstname||M.surname) As mem_name,
                    F.facid
        FROM Facilities As F
        Join Bookings As B
        On F.facid=B.facid
        Join Members As M
        On B.memid=M.memid
        Where M.memid!='0'
        )
SELECT F.name As fac_name, 
       sum(join_mem.membercost)+sum(join_guest.guestcost) As revenue
FROM Facilities AS F
LEFT JOIN join_guest ON F.facid = join_guest.facid
LEFT JOIN join_mem ON F.facid = join_mem.facid
GROUP BY F.name

"""
pd.read_sql(sql_stmt, conn)
```

**Q11**: Produce a report of members and who recommended them in alphabetic surname, firstname order

```python
sql_stmt="""
SELECT M1.surname, M1.firstname, M2.surname AS recomender
FROM Members AS M1
LEFT JOIN Members AS M2
ON M1.recommendedby=M2.memid
ORDER BY M1.surname, M1.firstname
"""
pd.read_sql(sql_stmt, conn)
```

**Q12**: Find the facilities with their usage by member, but not guests


```python
sql_stmt="""
SELECT DISTINCT F.name As fac_name, 
              (M.firstname||M.surname) As mem_name
FROM Facilities As F
Join Bookings As B
On F.facid=B.facid
Join Members As M
On B.memid=M.memid 
-- Exclude guests
WHERE M.memid!='0' 
"""
pd.read_sql(sql_stmt, conn)
```

**Q13**: Find the facilities usage by month, but not guests


```python
sql_stmt="""
SELECT DISTINCT F.name AS fac_name, 
              strftime('%m', B.starttime) AS usage_month
FROM Facilities As F
Join Bookings As B
On F.facid=B.facid
Join Members As M
On B.memid=M.memid 
-- Exclude guests
WHERE M.memid!='0' 
"""
pd.read_sql(sql_stmt, conn)
```
